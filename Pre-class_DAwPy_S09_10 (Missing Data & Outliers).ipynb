```python
df = sns.load_dataset("diamonds")
```

```python
df = df.select_dtypes(include = ['float64', 'int64'])
df
```

```python
df = df.dropna()
df
```

```python
df.isnull().sum()
```

```python
df_table = df["table"]
df_table
```

```python
q1 = df_table.quantile(0.25)
q3 = df_table.quantile(0.75)
IQR = q3-q1
lowerL = q1 - IQR*1.5
upperL = q3 + IQR*1.5
```

```python
df[(df.table < lowerL) | (df.table > upperL)]
```

```python
df[(df.table >= lowerL) & (df.table <= upperL)]
```
